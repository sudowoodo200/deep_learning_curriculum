### Visualizations

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
df = pd.read_csv("logs/output_metrics.csv")
```

```python
plot_df = df.pivot_table(index="model_size", columns="data_size", values="test_loss")
```

```python
sns.heatmap(plot_df, annot=True, fmt=".3f")
```
